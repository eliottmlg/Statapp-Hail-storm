```python
year = pd.get_dummies(dfreg['YEAR'], drop_first = True)
month = pd.get_dummies(dfreg['MONTH'], drop_first = True)
hour = pd.get_dummies(dfreg['HOUR'], drop_first = True)
latcat = pd.get_dummies(dfreg['LAT_CAT'], drop_first = True)
loncat = pd.get_dummies(dfreg['LON_CAT'], drop_first = True)
dfreg = pd.concat([dfreg, year, month, hour, latcat, loncat], axis = 1)
```
